&#9989; **Question 4.10 (4 point)**: Let's perform a GridSearch to see whether the accuracy can be improved. 

* Set param_grid to be `C`: 1e-2, 1e-1, 1e0, 1e1; `gamma`: 1e-4, 1e-3, 1e-2, 1e-1, 1e0; and `kernel`: linear, rbf.

````python
### Put your code here
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Step 1: Prepare data
df_numeric = numeric_features.copy()
le = LabelEncoder()
df_numeric['Species'] = le.fit_transform(df_numeric['Species'])

X = df_numeric.drop(columns='Species')
y = df_numeric['Species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Step 2: Set parameter grid
param_grid = {
    'C': [1e-2, 1e-1, 1e0, 1e1],
    'gamma': [1e-4, 1e-3, 1e-2, 1e-1, 1e0],
    'kernel': ['linear', 'rbf']
}

# Step 3: Perform grid search
grid_search = GridSearchCV(SVC(), param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Step 4: Evaluate best model
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# Print classification report
print("Best parameters:", grid_search.best_params_)
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Step 5: Plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le.classes_)
disp.plot(cmap='Blues')
plt.title("Confusion Matrix: Best SVM from GridSearch")
plt.show()
````

````python
import random 
import numpy as np
import math
import matplotlib.pyplot as plt
import time  
from IPython.display import display, clear_output
````

````python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from matplotlib.cbook import get_sample_data

def imscatter(x, y, image, ax=None, zoom=1):
    if ax is None:
        ax = plt.gca()
    try:
        image = plt.imread(image)
    except TypeError:
        # Likely already an array...
        pass
    im = OffsetImage(image, zoom=zoom)
    x, y = np.atleast_1d(x, y)
    artists = []
    for x0, y0 in zip(x, y):
        ab = AnnotationBbox(im, (x0, y0), xycoords='data', \
                            frameon=False)
        artists.append(ax.add_artist(ab))
    ax.update_datalim(np.column_stack([x, y]))
    ax.autoscale()
    return artists
````

**Human class**

````python
class human():
    
    ## initiate and parameters
    def __init__(self, roam_range, initial_x, initial_y, \
                 lower_bond, upper_bond, image_path):
        self.range = roam_range
        self.lw = lower_bond
        self.up = upper_bond
        self.image_path = image_path
        self.x = initial_x
        self.y = initial_y    

    ## draw images  
    def draw(self):
        imscatter(self.x, self.y, self.image_path, zoom=0.3)  
        
    ## random roaming  
    def roam(self):  
        self.dx = random.randint(-self.range, self.range)
        self.dy = random.randint(-self.range, self.range)
    
        # Clamp x and reverse direction if out of bounds
        if self.x <= self.lw:
            self.x = self.lw
            self.dx = abs(self.dx)
        elif self.x >= self.up:
            self.x = self.up
            self.dx = -abs(self.dx)
    
        # Clamp y and reverse direction if out of bounds
        if self.y <= self.lw:
            self.y = self.lw
            self.dy = abs(self.dy)
        elif self.y >= self.up:
            self.y = self.up
            self.dy = -abs(self.dy)
    
        self.x += self.dx
        self.y += self.dy        
````

**Zombie class** 

&#9989; **Question 5.2 (2 point)**:
Read through the code and explain what part I, part II, part III, and part IV are doing.

&#9989; **Do this**:
```
Part I gets the human's position so the zombie knows where to look.

Part II calculates the distance between the zombie and the human.

Part III checks if the human is close enough and, based on probability (kill_rate), decides whether the zombie catches (infects) them.

Part IV creates and returns a new zombie if an infection happens.
```

````python
class zombie():
    
    ## initiate and parameters
    def __init__(self, roam_range, hunt_radius, kill_rate, \
                 initial_x, initial_y, lower_bond, upper_bond, image_path):
        self.range = roam_range
        self.radius = hunt_radius
        self.kr = kill_rate
        self.x = initial_x
        self.y = initial_y
        self.lw = lower_bond
        self.up = upper_bond
        self.image_path = image_path

    ## draw images    
    def draw(self):
        imscatter(self.x, self.y, self.image_path, zoom=0.3)  
        
    ## random roaming   
    def roam(self):
        self.dx = random.randint(-self.range, self.range)
        self.dy = random.randint(-self.range, self.range)
    
        # Clamp x and reverse direction if out of bounds
        if self.x <= self.lw:
            self.x = self.lw
            self.dx = abs(self.dx)
        elif self.x >= self.up:
            self.x = self.up
            self.dx = -abs(self.dx)
    
        # Clamp y and reverse direction if out of bounds
        if self.y <= self.lw:
            self.y = self.lw
            self.dy = abs(self.dy)
        elif self.y >= self.up:
            self.y = self.up
            self.dy = -abs(self.dy)
    
        self.x += self.dx
        self.y += self.dy

    ## zombie hunts human
    def hunt(self, human):
        ## part I
        self.h_x = human.x
        self.h_y = human.y

        ## Part II
        dst = math.sqrt((self.x-self.h_x)**2 + (self.y-self.h_y)**2)
        
        catch = False 
        ## Part III
        if dst < self.radius:
            if random.random() < self.kr:
                catch = True
        return catch
        
    ## Part IV    
    def infect(self):
        new_zombie = zombie(self.range, self.radius, self.kr, \
                            self.x, self.y, self.lw, self.up, \
                            self.image_path) 
        return new_zombie
      
````
